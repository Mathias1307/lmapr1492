```python
s1 = SymmOp.from_xyz_string('-y,-x,z')

#Je choisis l'atome 
position = struct.sites[1].frac_coords
print("Position initiale de l'atome de Ti : " + str(position) + "\n")

#j'applique la symétrie
result = sym1.operate(position)
print("Position finale (après réflexion) de l'atome de Ti : " + str(result))

print("\nmatrice de réflexion associée: ")
print(sym1.rotation_matrix)

print("\nde translation associé:")
print(s1.translation_vector)

display(Image(filename='1.png'))
```
